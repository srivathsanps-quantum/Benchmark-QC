```python
import numpy as np
from ham_pyscf import H_gen

basis_input = "sto-6g"
elements = ["N", "N"]
ncas = 4 #Active orbitals 
nelecas = 4 #Active electrons
spin = 0
charge = 0


Rs = np.linspace(0.8, 2.0, 13)
for R in Rs:
    geom = f"""
    N  0.0  0.0  0.0
    N  0.0  0.0  {R}
    """
    H, E = H_gen(basis_input, ["N","N"], geom, spin, charge,
                 ncas, nelecas,
                 save=True, savefile="N2_PES_H.npz", geom_id=R)
```
