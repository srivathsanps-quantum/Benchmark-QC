```python
import numpy as np
from U2_ham import H_gen
import basis_set_exchange as bse

basis = bse.get_basis('ANO-RCC-MB', elements=['U'], fmt='nwchem')

basis_input = basis
elements = ["U"]
ncas = 6 #Active orbitals 
nelecas = 6 #Active electrons
spin = 0
charge = 0


Rs = (2.4, 2.48)
for R in Rs:
    geom = f"""
    U  0.0  0.0  0.0
    U  0.0  0.0  {R}
    """
    H, E = H_gen(basis_input, elements, geom, spin, charge,
                 ncas, nelecas,
                 save=True, savefile="U2_PES_H.npz", geom_id=R)
```
