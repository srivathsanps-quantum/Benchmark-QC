```python
import numpy as np
from U2_ham1 import H_gen, build_u2_reference
import basis_set_exchange as bse

basis = bse.get_basis('ANO-RCC-MB', elements=['U'], fmt='nwchem')

basis_input = basis
elements = ["U"]
ncas = 6 #Active orbitals 
nelecas = 6 #Active electrons
spin = 0
charge = 0

mol_ref, mo_ref = build_u2_reference(basis_input, ncas, nelecas)



Rs = (2.0, 2.2, 2.4, 2.48, 2.56, 2.6, 2.8)
for R in Rs:
    geom = f"""
    U  0.0  0.0  0.0
    U  0.0  0.0  {R}
    """
    H, E = H_gen(
        basis_input, elements, geom, spin, charge,
        ncas, nelecas,
        mol_ref, mo_ref,   # always the 2.5 Ã… reference
        save=True, savefile="U2_PES_H.npz", geom_id=R
    )
```
