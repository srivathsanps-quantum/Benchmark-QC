```python
import numpy as np
from FeS_pyscf import H_gen

basis_input = "ANO-RCC-MB"
elements = ["Fe", "S"]
ncas = 6 #Active orbitals 
nelecas = 6 #Active electrons
spin = 4 #No of unpaired electrons present
charge = 0


Rs = (1.826, 2.0, 2.2, 2.4, 2.6, 2.8, 3.0, 3.2, 3.6, 3.7, 4.0, 4.2, 4.5, 4.8)
for R in Rs:
    geom = f"""
    Fe  0.0  0.0  0.0
    S  0.0  0.0  {R}
    """
    H, E , op_pool= H_gen(basis_input, elements, geom, spin, charge,
                 ncas, nelecas,
                 save=True, savefile="FeS_PES_H.npz", geom_id=R)
```
